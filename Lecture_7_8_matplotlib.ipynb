## *Types of variables in statistics and data analysis*

Variables can be categorized into four main types 

- Nominal
- Ordinal
- Interval
- Ratio. 

Each type of variable represents different levels of measurement and has specific properties that affect how they can be analyzed and interpreted.

### Nominal Variable:

Nominal variables are categorical variables that represent different categories or groups without any inherent order or numerical significance. They are named because they allow us to assign names or labels to distinct groups. Examples of nominal variables include gender (e.g., male, female), marital status (e.g., single, married, divorced), and types of animals (e.g., dog, cat, bird).

**Key properties of nominal variables:**

- No inherent numerical value or order.
- Data can be represented using labels or names.
- Operations like **counting**, **mode**, and **frequency distributions** can be applied.
- We cannot take the mean, median, standard deviation or variance of nominal variables.

### Ordinal Variables:
Ordinal variables also represent categories, but they possess an inherent order or ranking. The differences between categories may not be quantifiable, but the order or sequence matters. Examples of ordinal variables include rating scales for satisfaction (e.g., very unsatisfied, unsatisfied, neutral, satisfied, very satisfied), educational levels (e.g., elementary, high school, bachelor's degree, master's degree), or rankings (e.g., 1st place, 2nd place, 3rd place).

**Example in detail**

The five richest Americans in the world. We know that the richest American holds the first position, followed by the second richest American, but we don’t know the exact difference in their wealth. We are unaware of whether the richest American earns just one dollar more or one billion dollars more than the second richest American. Another example is the class rankings, where students are categorized as first, second, and third positions: Excellent, Good, and Fair. **An ordinal variable provides information about the relative positions of these entities.** But we don't know the **exact differences in wealth or academic performance** between them.


**Key properties of ordinal variables:**

- Categories have a meaningful order.
- The distance between categories may not be uniform or measurable.
- Operations like counting, mode, frequency distributions, median, percentiles and ranking-based can be applied. 


### Interval Variables

Variables measured on either interval or ratio scales, in contrast to ordinal variable scale, contain information about both **relative value and distance between the categories**. **However, interval variables lack a true zero point**, meaning that a value of zero does not imply the absence of the variable being measured. An example of an interval variable is temperature measured in Celsius or Fahrenheit. A temperature of 0°C does not mean the absence of heat; it's just an arbitrary point on the scale.

**Key properties of interval variables:**

- Equal intervals between values.
- No true zero point.
- Basic arithmetic operations like addition and subtraction are meaningful.
- Mean, median, mode, standard deviation, variance, min, max etc. can be applied.

### Ratio Variables:

Ratio variables are similar to interval variables, but they do have a true zero point, which indicates the absence of the variable being measured. These variables have meaningful ratios between values, allowing for multiplication and division operations. Examples of ratio variables include height, weight, age, and income. For example, if I know that one member of my sample is 58 inches tall, another is 60 inches tall, and a third is 66 inches tall, I know who is the tallest and how much taller or shorter each member of my sample is in relation to the others. Because my height variable is measured using inches, and all inches are equal in length, the height variable is measured using a scale of equal intervals and provides information about both relative position and distance.

**Key properties of ratio variables**

- Equal intervals between values.
- A true zero point.
- All basic arithmetic operations are valid (addition, subtraction, multiplication, division).
- Mean, median, mode, standard deviation, variance, min, max etc. can be applied.

## Types of variables and appropriate visualizations

### Nominal Variables:
**Bar charts**: These are ideal for displaying the distribution of categories in nominal data. Each category is represented by a separate bar, and the height of the bar represents the frequency or count of each category.

**Pie charts**: These are suitable for showing the proportion or percentage distribution of different categories in nominal data.

### Ordinal Variables:
**Bar charts**: Similar to nominal variables, bar charts can effectively display the relative frequencies or counts of ordinal categories.

**Ordered bar charts**: These specialized bar charts arrange the bars in a specific order according to the ordinal ranking, emphasizing the ordinal nature of the data.

### Interval and Ratio Variables:

**Histograms:** Histograms are useful for displaying the *distribution of continuous data* **(single continuous variable), such as interval and ratio variables**. The data is grouped into intervals (bins), and the height of each bar represents the frequency or count within each interval.

**Box plots (Box-and-Whisker plots)**: *In boxplot, one variable is categorical (or even ordinal), and the other variable is numeric (continuous)*. We make boxplot if we are interested in seeing how the distribution of numeric variable varies with some
categorical variable.These plots summarize the distribution of data, including the median, quartiles, and outliers. They are suitable for interval and ratio variables and provide insights into the central tendency and variability of the data.

**Scatter plots:** Scatter plots are excellent for visualizing the relationship between two **continuous variables (interval or ratio)** and identifying patterns, trends, or correlations between them. Again, if you want to explore the joint distribution of two continuous variables, then scatter plot is the most useful visualization.

**Line charts:** Line charts are used to represent trends and patterns in time series data, which often involve interval or ratio variables.

## **Two Interfaces in Matplotlib:**

A potentially confusing feature of Matplotlib is its dual interfaces: 

- a convenient **MATLAB-style state-based interface**, 
- and a **more powerful object-oriented interface**

## *MATLAB-style code:*

Matplotlib was originally conceived as a Python alternative for MATLAB users, and much of its syntax reflects that fact. The MATLAB-style tools are contained in the pyplot (plt) interface. **It is important to recognize that this interface is stateful: it keeps track of the “current” figure and axes, which are where all plt commands are applied.** Matlab style coding is fast and convenient for simple visualization but it is difficult to track the **current axes and current figure** which we are working. Modifying the figures later on is also quite difficult.

## *OOP style matplotlib code:*

The object-oriented interface is available for these **more complicated situations, and for when you want more control over your figure**. Rather than depending on some notion of an “active” figure or axes, in the object-oriented interface the **plotting functions are "methods" of explicit Figure and Axes objects**.

## *Matlab-style coding*

```python
# Sample data (fake data)
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 8]

# Create a simple line plot
plt.plot(x, y, label='Data')

# Add labels and a legend
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Line Plot')
plt.text(x = 2.7, y = 7, s = "Minima")
plt.legend()

plt.savefig("simple_graph.png")

# Display the plot
plt.show()
```

In the above example, we use functions like **plt.plot(), plt.xlabel(), plt.ylabel(), plt.title(), and plt.legend()** to create and customize the plot. The plt prefix is used to call functions directly from the matplotlib.pyplot module.

## Object-oriented matplotlib style coding:
This style involves **creating explicit figure and axes objects**, and you interact with these objects to create and customize the plots. It provides more control and flexibility, making it easier to create complex plots or multiple subplots

```python
# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 8]

# Create a figure and an Axes object
fig, ax = plt.subplots()

# Create the line plot using the plot() method of the Axes object
ax.plot(x, y, label='Data')

# Add labels and a legend using methods of the Axes object
ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')
ax.set_title('Line Plot')
ax.legend()

# Display the plot
plt.show()
```

In the above example, we first create a figure and an Axes object using plt.subplots(). The fig variable represents the figure, and the ax variable represents the Axes object.

**We then use the plot() method of the ax object to create the line plot. The set_xlabel(), set_ylabel(), set_title(), and legend() methods are used to customize the plot properties.**

The *key difference here is that we use the ax object to access methods for creating and customizing the plot, rather than calling functions directly from the matplotlib.pyplot module*.

# **Flights dataset**

```python
flights = pd.read_csv("flights_Data.csv")
```

```python
flights.columns
```

## *Histogram*

- visualizes the distribution of a single continous variable

```python
flights_distance = flights["distance"].tolist()
flights_distance
```

```python
plt.hist(flights_distance, bins = 20, edgecolor = "blue", linewidth =3, color = "green", log = True) 

plt.xlabel("Flights distance")
plt.ylabel("flights frequency")
plt.title("flights distance vs flights frequency")
plt.show()
```

## *Barplot*

```python
flights.origin.unique()
```

```python
origin_airports = flights.groupby(["origin"]).count()["flight"].index.tolist()

print(origin_airports)

count_flights =  flights.groupby(["origin"]).count()["flight"].tolist()

print(count_flights)
```

```python
plt.bar(origin_airports, count_flights)

plt.xlabel("Origin airports")
plt.ylabel("flights frequency")
plt.title("Busy origin airports")
plt.show()
```

## *Pie chart*

```python
plt.pie(count_flights, labels = origin_airports, colors = ["Green","red", "blue"],
        wedgeprops = {"edgecolor": "black"},
        explode = [0, 0, 0.10],
        autopct = "%1.1f%%")

plt.title("Busiest origin airports")
```

## *Barchart: Average flight delay by origin airports*

```python
origin_airport = flights.groupby(["origin"])["dep_delay"].mean().index.tolist()
print(origin_airport)

avg_delay = flights.groupby(["origin"])["dep_delay"].mean().tolist()
avg_delay = [round(x,2) for x in avg_delay]

print(avg_delay)
```

```python
plt.bar(origin_airport, avg_delay)
plt.xlabel("origin airports")
plt.ylabel("Average delay")
plt.title("Average flight delays on different origin airports")
```

## *Bar chart: Top ten airlines*

```python
top_ten_carriers = flights.groupby("carrier").count()["flight"].sort_values(ascending = False)[0:10].index.tolist()
```

```python
flights_operated = flights.groupby("carrier").count()["flight"].sort_values(ascending = False)[0:10].tolist()
```

```python
plt.barh(top_ten_carriers, flights_operated)
plt.title("Top ten airlines in USA in 2013")
```
